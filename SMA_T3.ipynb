```python
import pandas as pd
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

# Specify the path to your CSV file
filename = 'matcha.csv'
file_path = f"tweets-data/{filename}"

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path, delimiter=",")

STOPWORDS = set(stopwords.words('english'))

def clean_text(text):
    text = text.lower()
    text = re.sub(r'http\S+', '', text)      # hapus url
    text = re.sub(r'@\w+', '', text)         # hapus mention
    text = re.sub(r'#', '', text)            # hapus tanda #
    text = re.sub(r'[^a-z\s]', ' ', text)    # hapus simbol & angka
    tokens = [t for t in text.split() if t not in STOPWORDS]
    return " ".join(tokens)

df['clean'] = df['full_text'].apply(clean_text)

display(df.head())
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

def get_sentiment(text):
    score = analyzer.polarity_scores(text)['compound']
    if score >= 0.05:
        return "Positive"
    elif score <= -0.05:
        return "Negative"
    else:
        return "Neutral"

df['sentiment'] = df['clean'].apply(get_sentiment)
df.to_csv("matcha_sentiment.csv", index=False, encoding="utf-8")
display(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,6))
df['sentiment'].value_counts().plot.pie(autopct='%1.1f%%', startangle=140)
plt.title("Distribusi Sentimen matcha")
plt.ylabel("")
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='sentiment', order=['Positive','Neutral','Negative'])
plt.title("Jumlah Tweet per Kategori")
plt.show()
```
